```python
energy_copy = energy_data.copy()
```

```python
energy_copy["day"] = energy_data["Datetime"].dt.day
energy_copy["month"] = energy_data["Datetime"].dt.month
```

```python
energy_grouped = energy_copy.groupby(["day","month"], as_index = False).sum()
```

```python
energy_grouped["date"] = energy_grouped["day"].astype(str) + energy_grouped["month"].astype(str)
```

```python
plt.scatter(energy_grouped["date"],energy_grouped["Wind_Gen_MW"], color = 'red')
plt.title("Yearly Wind Energy Generation") #title
plt.xlabel("Date") #x label
plt.ylabel("Energy MW") #y label
```

```python
plt.scatter(energy_grouped["date"],energy_grouped["Solar_Gen_MW"], color = 'orange')
plt.title("Yearly Solar Energy Generation") #title
plt.xlabel("Date") #x label
plt.ylabel("Energy MW") #y label
```
